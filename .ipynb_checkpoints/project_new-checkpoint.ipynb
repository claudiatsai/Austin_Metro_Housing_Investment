### Reshape from Wide to Long Format

```python
def melt_data(df):
    ''' 
        Takes a dataframe with datetime data that is in wide format and melts it into long format;      
        Transforms data into datetime object with time as index, defaulted as monthly.                     

        df          Dataframe to transform     
    '''
    
    melted = pd.melt(df, id_vars=['RegionID', 'Zipcode', 'City', 'State', 'Metro', 'CountyName', 'SizeRank'], 
                      var_name='Month', value_name='MeanValue')
    melted['Month'] = pd.to_datetime(melted['Month'], format='%Y-%m')
    melted = melted.dropna(subset=['MeanValue'])
    return melted
#     return melted.groupby('time').aggregate({'value':'mean'})
```

### Make the datetime as index

```python
def make_datetime_index(df,col_to_make_index = 'Month', drop=False, verbose=True):
    '''           
         Transforms data into datetime object with time as index, defaulted as monthly     
         
         df          Dataframe to transform     
    '''

    df[col_to_make_index] = pd.to_datetime(df[col_to_make_index], errors='coerce')
    df = df.set_index('Month', drop=False)
    if verbose:
        display(df.index)
    return df
```

```python
def check_stationarity(ts):
    roll_mean = ts.rolling(window=8, center=False).mean()
    roll_std = ts.rolling(window=8, center=False).std()

    fig = plt.figure(figsize=(12,7))
    plt.plot(ts, color='green', label='Original')
    plt.plot(roll_mean, color='blue', label='Rolling Mean')
    plt.plot(roll_std, color='red', label = 'Rolling STD')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    dftest = adfuller(ts)

    # Extract and display test results in a user friendly manner
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
        
    print ('Results of Dickey-Fuller test: \n')

    print(dfoutput)
    
    
    return None
```

```python
def acf_pacf(df,alags=30,plags=30):
    ''' 
        Performs acf/pacf results plot      
                             
        df          Dataframe to Analyse     
    '''

    #Create figure
    fig,(ax1,ax2) = plt.subplots(2,1,figsize=(14,12))
    #Make ACF plot
    plot_acf(df,lags=alags, zero=False,ax=ax1)
    #Make PACF plot
    plot_pacf(df,lags=plags, ax=ax2)
    plt.show()
```

```python
def AIC_PDQS(df):   
    ''' 
         Runs grid search to return lowest AIC result for permutataions of pdq/s values in range 0,2  
                          
         df          Dataframe to anlyse for best pdq/s permutation    
    '''
    
    # Define the p, d and q parameters to take any value between 0 and 2
    p = d = q = range(0, 2)
    
#     Auto-Regressive (p) -> Number of autoregressive terms.
#     Integrated (d) -> Number of nonseasonal differences needed for stationarity.
#     Moving Average (q) -> Number of lagged forecast errors in the prediction equation.

    # Generate all different combinations of p, q and q triplets
    pdq = list(itertools.product(p, d, q))

    # Generate all different combinations of seasonal p, q and q triplets
    pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]


    # Run a grid with pdq and seasonal pdq parameters calculated above and get the best AIC value
    ans = []
    for comb in pdq:
        for combs in pdqs:
            try:
                mod = sm.tsa.statespace.SARIMAX(df,
                                            order=comb,
                                            seasonal_order=combs,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

                output = mod.fit()
                ans.append([comb, combs, output.aic])
                print('ARIMA {} x {}12 : AIC Calculated ={}'.format(comb, combs, output.aic))
            except:
                continue
            
    # Find the parameters with minimal AIC value
    ans_df = pd.DataFrame(ans, columns=['pdq', 'pdqs', 'aic'])
    print(ans_df.loc[ans_df['aic'].idxmin()])
    return ans_df
```

## Fitting an ARIMA Time Series Model

```python
# Train, test split
def train_test(df):
   
    #Use data before 2017-05 as training data set 
    train = df[:'2017-04']
    #Use 12 month prior to 2018-04 as test data set test, which starts from 2017-05
    test = df['2017-05':]
    return train, test
```

```python
def model_fit(df,pdq=(1,0,1),pdqs=(0,0,0,1)):
    ''' 
        Performs SARAMA model given defined pdq/s values and returns results     
        
        df          Dataframe to Analyse   
        
        Returns train/test df on the same df modelled   
    '''

    train, test = train_test(df)
    model = SARIMAX(train,order=pdq,seasonal_order=pdqs, 
                                        )
    results = model.fit()
    results.summary
    residuals = results.resid
    print(results.summary())
    results.plot_diagnostics(figsize=(14,14))
    plt.show();
    return train, test, results


#print(output.summary().tables[1])
```

```python
#Fit the SARIMA model and get results.
pdq = (0, 1, 1)
pdqs = (1, 1, 1, 12)

train, test, results = model_fit(ts_diff_78702,pdq=pdq,pdqs=pdqs)
```

```python
def test_model(df,pdq=(1,0,1),pdqs=(0,0,0,1), display=True):
    ''' 
            Predicts for 12 months on test model with SARIMA model  
                      
            df          Dataframe to perform predictions    
    '''

    X = df.values
    train, test = X[:-12],X[-12:]
    history = [x for x in train]
    predictions = []
    for t in range(len(test)):
        model = SARIMAX(history, order=pdq,seasonal_order=pdqs, enforce_stationarity=False, 
                                        enforce_invertibility=False)
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        history.append(test[t])
    #rmse = sqrt(MSE(test, predictions))
    #print('SARIMA model RMSE on test data: %.5f' % rmse)
    #print('SARIMA model RMSE on test data')
    if display:
        plt.figure(figsize=(13,6))
        plt.title('Actual Test Data vs. Predictions')
        plt.plot(history[-12:],label='Actual', color='b')
        plt.plot(predictions,label='Predictions',color='r')
        plt.legend(loc='best')
        plt.show()
```

```python
def train_model(train, results, display = True):
    ''' 
        Returns train versus test predictions pefromed on on SARIMA model    
        Transforms data into datetime object with time as index.                     

        train          Train Dataframe for period defined
        results        Results from SARIMA model
    '''

    train_pred = results.predict(-12)
#     rmse = sqrt(MSE(train[-12:],train_pred))
#     print(f'SARIMA model RMSE on train data: %.5f' % rmse)
#    print(f'SARIMA model RMSE on train data')
    if display:
        plt.figure(figsize=(13,6))
        train[-60:].plot(label='Actual',color='b')
        train_pred.plot(label='Predicted',color='r')
        plt.legend(loc='best')
        plt.title('Actual Train Data vs. Predicted Returns')
        plt.show()
```

## Validating the Model

```python
def forecast_model(df,pdq=(1,0,1),pdqs=(0,0,0,1), display=True,zc='input zipcode'):
    
    ''' 
        Performs predictions on SARIMA model up to 10 years in the future   
                           
         df          Dataframe to model    
    '''

    #model = SARIMAX(df, order=pdq,seasonal_order=pdqs, enforce_stationarity=False, 
                                        #enforce_invertibility=False)
    model = SARIMAX(df, order=pdq,seasonal_order=pdqs)

    model_fit = model.fit()
    output = model_fit.get_prediction(start='2017-05',end='2027-06', dynamic=True)
    forecast_ci = output.conf_int()
    if display:
        fig, ax = plt.subplots(figsize=(13,6))
        output.predicted_mean.plot(label='Forecast')
        ax.fill_between(forecast_ci.index,forecast_ci.iloc[:, 0],forecast_ci.iloc[:, 1],
                        color='k', alpha=.25,label='Conf Interval')
        plt.title('Forecast of Monthly Predicted Property Value')
        plt.xlabel('Time')
        plt.legend(loc='best')
        plt.show()

        year_0 = output.predicted_mean['2017-05-01']
        year_1= output.predicted_mean['2018-05-01']
        year_1_ROI_100K = ((year_1 - year_0)*100)/year_0
        year_3= output.predicted_mean['2020-05-01']
        year_3_ROI_100K = ((year_3 - year_0)*100)/year_0
        year_5= output.predicted_mean['2022-05-01']
        year_5_ROI_100K = ((year_5 - year_0)*100)/year_0
        year_10= output.predicted_mean['2027-05-01']
        year_10_ROI_100K = ((year_10 - year_0)*100)/year_0
        
        print(f'Predicted property mean value at initial investment date 2017-05: {round(year_0,2)}')
        #print(f'Total expected return in 1 year: {round(year_1*100,2)}%')
        print(f'Predicted property mean value in 1 year: {round(year_1,2)} is {round(year_1_ROI_100K,2)} %ROI')
        print(f'Predicted property mean value in 3 years: {round(year_3,2)} is {round(year_3_ROI_100K,2)} %ROI')
        print(f'Predicted property mean value in 5 years: {round(year_5,2)} is {round(year_5_ROI_100K,2)} %ROI')
        print(f'Predicted property mean value in 10 years: {round(year_10,2)} is {round(year_10_ROI_100K,2)} %ROI')

        tot_ret = [zc,year_1,year_3,year_5,year_10]
    return tot_ret
```

Overall country data, by city, state or region code
At different time periods including the full time period

The monthly rolling mean period can be extended to n days, n months, n years, depending on the amount of data being analysed versus prediction power, in order to further remove trends and increase stationarity of data.
